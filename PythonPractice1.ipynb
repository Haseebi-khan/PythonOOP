**Operators in Python**

```python
a, b = 15, 4
print(f"a + b = {a + b}")       # 19
print(f"a - b = {a - b}")       # 11
print(f"a * b = {a * b}")       # 60
print(f"a / b = {a / b}")       # 3.75
print(f"a // b = {a // b}")     # 3 (floor division)
print(f"a % b = {a % b}")       # 3 (remainder)
print(f"a ** b = {a ** b}")     # 15^4 = 50625
```

```python
m, n = 5, 3
print(m & n)   # 1 (0101 & 0011 = 0001)
print(m | n)   # 7 (0101 | 0011 = 0111)
print(m ^ n)   # 6 (0101 ^ 0011 = 0110)
print(~m)      # -6 (inverts bits: 0101 → ...1010 = -6 in two's complement)
print(m << 1)  # 10 (shift left: 0101 → 1010)
print(m >> 1)  # 2  (shift right: 0101 → 0010)
```
