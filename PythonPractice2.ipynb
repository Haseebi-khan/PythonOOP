```python
# Determine if a year is a leap year
year = int(input("Enter a year: "))

if year % 400 == 0:
    print("Leap year")
elif year % 4 == 0:
    if year % 100 != 0:
        print("Leap year")
    else:
        print("Not a leap year")      # Indentation issue
else:
    print("Not a leap year")

# Ternary operator misuse
age = 25
message = "Senior" if age >= 65 else "Adult" if age >= 18 else "Child" if age >= 13 else "Minor"
print(message)  # Hard to read, but is it correct? Let's find the actual error
```

# **Mini-Project**
Write a short script that:

Asks the user for their age (as an integer).

Asks if they have a valid passport (input "yes" or "no").

Asks if they have a visa (if required based on age and travel plans – you can simplify: for this exercise, visa is needed if age < 18 or age > 60).

Uses if-elif-else to determine and print:

If age < 18: "You need parental consent to travel."

If age ≥ 18 and passport = "yes" and visa = "yes": "You are ready to travel!"

If age ≥ 18 and (passport = "no" or visa = "no"): "You need to complete your travel documents."

If age > 60: "Please consult your doctor before traveling."

Use a ternary operator to set a variable travel_status to "Eligible" if all conditions are met, else "Not eligible", and print it.

Test with different inputs to verify all branches work. This will solidify your understanding of conditional logic.

```python
age = int(input("Enter your age: "))
passport = str(input("Do you have valid passport ('yes' or 'no'): "))
visa = str(input("Do you visa: ('yes' or 'no'): "))

if visa.lower() == "yes":
    if age < 18:
        print("You need parental consent to travel.")
    elif age >= 18 and passport.lower() == "yes" and visa.lower() == "yes":
        print("You are ready to travel!")
    elif age>= 18 and passport.lower() == "no" or visa.lower() == "no": 
        print("You need to complete your travel documents.")
    elif age > 60: 
        print("Please consult your doctor before traveling.")
else:
    print("Complete your visa process.")

travel_status = "Eligible" if (age >= 18 and age <= 60 and ( passport.lower() == "yes" and visa.lower() == "yes" )) else "Not Eligible"
print("travel_status: ", travel_status)
```
