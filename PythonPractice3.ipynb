````python
class Teacher:
    def __init__(self):
        self.count = 0
        self.teacher_names = []
        self.teacher_id = []
        self.salary = []
        self.email = []


class Management:
    def __init__(self):
        self.count = 0
        self.manage_names = []
        self.manage_id = []
        self.salary = []


class Student:
    def __init__(self):
        self.count = 0
        self.std_name = []
        self.std_id = []


class Person_info:
    def __init__(self):
        self.__t = Teacher()
        self.__m = Management()
        self.__s = Student()

    def data_entry(self):
        p_type = input("Student / Teacher / Management: ").lower()

        if p_type == "student":
            if self.__s.count < 5:
                name = input("Enter Name: ")
                self.__s.std_name.append(name)
                self.__s.std_id.append(self.__s.count)
                self.__s.count += 1

        elif p_type == "teacher":
            if self.__t.count < 5:
                name = input("Enter Name: ")
                salary = input("Enter Salary: ")
                email = input("Enter Email: ")

                if not email.endswith("@gmail.com"):
                    print("Invalid Email!")
                    return

                self.__t.teacher_names.append(name)
                self.__t.teacher_id.append(self.__t.count)
                self.__t.salary.append(salary)
                self.__t.email.append(email)

                self.__t.count += 1

        elif p_type == "management":
            if self.__m.count < 5:
                name = input("Enter Name: ")
                salary = input("Enter Salary: ")

                self.__m.manage_names.append(name)
                self.__m.manage_id.append(self.__m.count)
                self.__m.salary.append(salary)

                self.__m.count += 1

    def display(self):
        p_type = input("Student / Teacher / Management: ").lower()
        search_id = int(input("Enter ID: "))

        if p_type == "student":
            if search_id in self.__s.std_id:
                i = self.__s.std_id.index(search_id)
                print(self.__s.std_name[i], search_id)

        elif p_type == "teacher":
            if search_id in self.__t.teacher_id:
                i = self.__t.teacher_id.index(search_id)
                print(self.__t.teacher_names[i])
                print(self.__t.email[i])

        elif p_type == "management":
            if search_id in self.__m.manage_id:
                i = self.__m.manage_id.index(search_id)
                print(self.__m.manage_names[i])


def main():
    person = Person_info()

    while True:
        choice = int(input("1-Add 2-Display 0-Exit: "))

        if choice == 1:
            person.data_entry()
        elif choice == 2:
            person.display()
        elif choice == 0:
            print("Bye!")
            break


if __name__ == "__main__":
    main()
````

````python
class Binary:
    def __init__(self):
        self.__string = ""
    
    def dataEntry(self):
        self.__string = input("Enter the Binary string: ")
    
    def display(self):
        print(f"The Binary String is: {self.__string}")
    
    def check_Binary(self):
        
        found = True
        
        for i in self.__string:
            if i != '0' and i != '1':
                found = False
                break
        
        return found
    
    def show_Error(self):
        print("String is not a binary string: ", self.__string)
        
    def ones_Compliment(self):
            compliment_str = ""

            for i in self.__string:
                if i == '1':
                    compliment_str += '0'
                else:
                    compliment_str += '1'

            print(f"The One's Complement is: {compliment_str}")
            
def main():
    s = Binary()
    s.dataEntry()
    
    if s.check_Binary() == False:
        s.show_Error()
    else:
        s.display()
        s.ones_Compliment()
    
if __name__ == "__main__":
    main()
````

````python
def greatestIndex(lst: list):
    
    hightestIndexValue = ''
    
    if lst[0] > lst[1] and lst[0] > lst[2]:
        hightestIndex = "First Index Value is highest. "
    elif lst[1] > lst[0] and lst[1] > lst[2]:
        hightestIndex = "2nd Index Value is highest. "
    elif lst[2] > lst[0] and lst[2] > lst[1]:
        hightestIndex = "3rd Index Value is highest. ";
    
    return hightestIndex 

def hightestValue(lst: list):
    
    MAX = lst[0]
    
    for i in lst:
        if i > MAX:
            MAX = i
    
    return MAX

def reverse(num: int):

    last = num % 10
    seclast = (num / 10) % 10
    sec = (num / 100) % 10
    one = (num / 1000)

    print(f" {last} << {seclast} << {sec} <<  {one}.")

def main():
    import random

    number_arr = []

    random.seed(42)

    for i in range(3):
        number_arr.append(random.randint(1,1800))
        
    for i in number_arr:
        print(i, end=" ")
        
    print()
    index = greatestIndex(number_arr)
    Value = hightestValue(number_arr)

    print(f"{index}\nValue is: {Value}.")
    
    print("\nReverse Number: ")
    reverse(Value)
    

if __name__ == '__main__':
    main()
````

````python
import copy

class A:
    def __init__(self, v='ss'):
        self.value = v
        self.nested = [1, 2, 3]  # For demonstrating deep copy
    
    def __copy__(self):
        """Shallow copy - creates new instance with copied value"""
        print("__copy__ called")
        p = A(self.value)
        p.nested = self.nested.copy()  # Shallow copy of nested list
        return p
    
    def __deepcopy__(self, memo=None):
        """Deep copy - recursively copies all nested objects"""
        print("__deepcopy__ called")
        if memo is None:
            memo = {}
        
        # Create new instance
        h = A(self.value)
        
        # Deep copy the nested list
        h.nested = copy.deepcopy(self.nested, memo)
        return h

def main():
    
    # Create original object
    original = A("original object")
    original.nested.append(99)
    print(f"Original value: {original.value}")
    print(f"Original nested: {original.nested}")
    print(f"Original ID: {id(original)}")
    # above Obj go to garbage.
    
    
    # Create objects
    original = A("test")
    original.nested.append(100)

    # This will call __deepcopy__ automatically
    deep_copied = copy.deepcopy(original)

    print(f"Original nested: {original.nested}")
    print(f"Deep copied nested: {deep_copied.nested}") 
    
    print("\nInstance IDs:")
    print(f"Original ID: {id(original)}")
    print(f"Deep Copy ID: {id(deep_copied)}\n")
    
    # Change in Original must Not Reflect in Deep Copy
    deep_copied.nested.pop()
    deep_copied.nested.pop()
    deep_copied.nested.pop()
    print(f"Original value: {original.value}")
    print(f"Original nested: {original.nested}")
    print(f"Deep Copy value: {deep_copied.value}")
    print(f"Deep Copy nested: {deep_copied.nested}\n")
    
    
        
    original.nested = [1, 2, 3, 100]

    # Shallow copy (SHARES nested list)
    shallow_copied = copy.copy(original)
    shallow_copied.nested.append(999)

    print(f"Original nested: {original.nested}")      # [1, 2, 3, 100, 999]
    print(f"Shallow copy nested: {shallow_copied.nested}")  # [1, 2, 3, 100, 999]
    print(f"Changes DO reflect - They SHARE the list!\n")
    
if __name__ == "__main__":
    main()
````

````python
import copy

class Line:
    def __init__(self, line):
        self.__ptr = line # int is immutable
        self.__shared_list = [1, 2, 3]  
        
    def setter(self, x):
        self.__ptr = x
        
    def modify_shared_list(self, value):
        self.__shared_list.append(value)
        
    def __copy__(self):
        obj = Line(self.__ptr)
        obj.__shared_list = self.__shared_list  
        return obj
    
    def __deepcopy__(self, memo=None):
        if memo is None:
            memo = {}
        obj = Line(self.__ptr)

        obj.__shared_list = copy.deepcopy(self.__shared_list, memo)
        return obj
    
    def display(self):
        print(f"Ptr: {self.__ptr}, Shared List: {self.__shared_list}")
        
def main():
    
    print("=== DEMONSTRATING THE REAL DIFFERENCE ===\n")
    
    obj1 = Line(19)
    obj1.modify_shared_list(100)
    print("Original:", end=" ")
    obj1.display()
    
    print("\n--- Creating SHALLOW copy ---")
    obj2 = copy.copy(obj1)
    
    print("After shallow copy:")
    print("  Original:", end=" ")
    obj1.display()
    print("  Shallow:", end=" ")
    obj2.display()
    
    
    # Modify shared list through obj2
    print("\n--- Modifying shallow copy's list ---")
    obj2.modify_shared_list(999)
    
    print("  Original (AFFECTED!):", end=" ")
    obj1.display()
    print("  Shallow:", end=" ")
    obj2.display()
    
    # DEEP COPY - INDEPENDENT list
    print("\n--- Creating DEEP copy from original ---")
    obj3 = copy.deepcopy(obj1)
    
    print("After deep copy:")
    print("  Original:", end=" ")
    obj1.display()
    print("  Deep:", end=" ")
    obj3.display()
    
    # Modify deep copy's list
    print("\n--- Modifying deep copy's list ---")
    obj3.modify_shared_list(888)
    print("  Original (UNCHANGED!):", end=" ")
    obj1.display()
    print("  Deep:", end=" ")
    obj3.display()
    
    # Show that obj2 and obj3 are different
    print("\n--- Independent changes to shallow copy ---")
    obj2.setter(55)
    obj3.setter(66)
    print("  Original:", end=" ")
    obj1.display()
    print("  Shallow (ptr changed, list shared):", end=" ")
    obj2.display()
    print("  Deep (completely independent):", end=" ")
    obj3.display()
    
if __name__ == "__main__":
    main()
        
````

````python
import copy

class Customer:
    def __init__(self):
        self.name = ''
        self.address = [' '] * 10
    
    def __deepcopy__(self, memo = None):
        if memo is None:
            memo = {}
        
        obj = Customer()
        obj.name = self.name
        obj.address = self.address.copy()        
        memo[id(self)] = obj
        
        return obj
        
    def __copy__(self):
        obj = Customer()
        obj.name = self.name
        obj.address = self.address
        return obj
    
    def setter(self, name:str, address:list):
        self.name = name
        for i in range(min(len(address), 10)):
            self.address[i] = address[i]
        
    def getter(self):
        print(f"Name: {self.name}")
        address_str = ''.join(self.address).rstrip()
        print(f"Address: {address_str}")
        
    def modify_address(self, new_address):
        for i in range(min(len(new_address), 10)):
            self.address[i] = new_address[i]

    
def main():
    c1 = Customer()
    
    address :list = []
    a = "Street asdf"
    
    for i in range(len(a)):
        address.append(a[i])
    
    c1.setter("Khan", address)

    print("Original Customer:")
    c1.getter()
    
    print("\n" + "="*50)
    print("=== Testing Shallow Copy ===\n")
    
    # Create shallow copy
    c2 = copy.copy(c1)
    print("\nAfter shallow copy:")
    print("Original:", end=" ")
    c1.getter()
    print("Shallow :", end=" ")
    c2.getter()
    
    # Modify shallow copy's address (will affect original)
    print("\n--- Modifying shallow copy's address ---")
    c2.modify_address(list("NewAddrX"))
    print("After modification:")
    print("Original (AFFECTED!):", end=" ")
    c1.getter()
    print("Shallow copy:", end=" ")
    c2.getter()
    
    print("\n" + "="*50)
    print("=== Testing Deep Copy ===\n")
    
    # Create deep copy from a fresh object
    c3 = Customer()
    c3.setter("Khan", address)
    
    c4 = copy.deepcopy(c3)
    print("\nAfter deep copy:")
    print("Original:", end=" ")
    c3.getter()
    print("Deep    :", end=" ")
    c4.getter()
    
    # Modify deep copy's address (will NOT affect original)
    print("\n--- Modifying deep copy's address ---")
    c4.modify_address(list("NewAddrY"))
    print("After modification:")
    print("Original (UNCHANGED!):", end=" ")
    c3.getter()
    print("Deep copy:", end=" ")
    c4.getter()
    
    # Verify independence
    print("\n" + "="*50)
    print("=== Verification ===")
    print(f"c1 and c2 share address list? {c1.address is c2.address}")  # True
    print(f"c3 and c4 share address list? {c3.address is c4.address}")  # False
    print(f"c1 and c2 are same object? {c1 is c2}")  # False
    print(f"c3 and c4 are same object? {c3 is c4}")  # False
        

if __name__ == "__main__":
    main()
        
````

````python
class Stock:
    
    no_of_stocks = 0
    
    def __init__(self):
        self.__ticketsymbol = [''] * 5 
        self.__companyname = ''
        self.__currentprice = 0.0
        self.__historicalprices = []
        Stock.no_of_stocks += 1
    
    def setter(self, companyname, currentprice, historicalprices, ticketsymbol: None):
        
        if ticketsymbol is not None:
            if len(ticketsymbol) == 5:
                self.__ticketsymbol = ticketsymbol
            else:
                print(f"Warning: Ticket symbol must be 5 characters. Got {len(ticketsymbol)}")
                if len(ticketsymbol) > 5:
                    self.__ticketsymbol = list(ticketsymbol[:5])
                else:
                    self.__ticketsymbol = list(ticketsymbol) + [''] * (5 - len(ticketsymbol))
                    
        self.__companyname = companyname
        self.__currentprice = currentprice
        
        self.__historicalprices = []
        
        for old_price in historicalprices:
            if isinstance(old_price, (int, float)):  # Fixed syntax
                self.__historicalprices.append(float(old_price))  # Convert to float
            else:
                raise ValueError("Historical Prices must be integer or floating points.")
        
    
    def get_ticket_symbol(self):
        return ''.join(self.__ticketsymbol).rstrip()
    
    def get_company_name(self):
        return self.__companyname
    
    def get_current_price(self):
        return self.__currentprice
    
    def get_historical_prices(self):
        return self.__historicalprices.copy()
    
    def get_average_price(self):
        if len(self.__historicalprices) == 0:
            return self.__currentprice
        
        return sum(self.__historicalprices) / len(self.__historicalprices)
        
    def get_overvalued(self):
        average = self.get_average_price()
        
        if self.__currentprice > average:
            print("Stock is currently overvalued.")
        else:
            print("Stock is not overvalued.")
            
    def display(self):
        """Display all stock information"""
        print("\n" + "="*50)
        print(f"STOCK INFORMATION")
        print("="*50)
        print(f"Ticket Symbol: {''.join(self.__ticketsymbol).rstrip()}")
        print(f"Company Name: {self.__companyname}")
        print(f"Current Price: ${self.__currentprice:.2f}")
        print(f"Historical Prices: {self.__historicalprices}")
        print(f"Total Stocks Created: {Stock.no_of_stocks}")
        print("="*50)
        
        
def main():
    
    historial_prices_list= [32, 75, 67, 83, 12]
    
    tesla = Stock();
    
    tesla.setter("Tesla, Inc.", 50, historial_prices_list, "TSLA1")
    
    tesla.display()

    pass

if __name__ == "__main__":
    main()
````
